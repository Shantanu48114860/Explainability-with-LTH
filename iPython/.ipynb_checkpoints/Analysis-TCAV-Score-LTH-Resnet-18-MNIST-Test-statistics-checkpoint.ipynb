```python
def get_TCAV(cav_flattening_type):
    print(f"cav_flattening type: {cav_flattening_type}")
    TCAV = []
    
    concept_zero_class_even = []
    concept_one_class_even = []
    concept_two_class_even = []
    concept_three_class_even = []
    concept_four_class_even = []
    concept_five_class_even = []
    concept_six_class_even = []
    concept_seven_class_even = []
    concept_eight_class_even = []
    concept_nine_class_even = []
    
    concept_zero_class_odd = []
    concept_one_class_odd = []
    concept_two_class_odd = []
    concept_three_class_odd = []
    concept_four_class_odd = []
    concept_five_class_odd = []
    concept_six_class_odd = []
    concept_seven_class_odd = []
    concept_eight_class_odd = []
    concept_nine_class_odd = []
    
    
    TCAV_path = os.path.join(
        logs,
        "predictions",
        "prune-statistics",
        model_arch,
        dataset_name,
        f"Prune_type_{prune_type}",
        f"cav_flattening_type_{cav_flattening_type}"
        "TCAV-scores"
    )
    print(TCAV_path)
    for ite in range(35):
        tcav_file_name = f"TCAV_scores_file_pruning_iter_{ite}.pkl"
        tcav_file = open(
                os.path.join(TCAV_path, tcav_file_name),
                "rb")
        TCAV.append(pickle.load(tcav_file))
    
    for ite in range(35):
        concept_zero_class_even.append(TCAV[ite]['Zero']['class_0'])
        concept_one_class_even.append(TCAV[ite]['One']['class_0'])
        concept_two_class_even.append(TCAV[ite]['Two']['class_0'])
        concept_three_class_even.append(TCAV[ite]['Three']['class_0'])
        concept_four_class_even.append(TCAV[ite]['Four']['class_0'])
        concept_five_class_even.append(TCAV[ite]['Five']['class_0'])
        concept_six_class_even.append(TCAV[ite]['Six']['class_0'])
        concept_seven_class_even.append(TCAV[ite]['Seven']['class_0'])
        concept_eight_class_even.append(TCAV[ite]['Eight']['class_0'])
        concept_nine_class_even.append(TCAV[ite]['Nine']['class_0'])
        
        concept_zero_class_odd.append(TCAV[ite]['Zero']['class_1'])
        concept_one_class_odd.append(TCAV[ite]['One']['class_1'])
        concept_two_class_odd.append(TCAV[ite]['Two']['class_1'])
        concept_three_class_odd.append(TCAV[ite]['Three']['class_1'])
        concept_four_class_odd.append(TCAV[ite]['Four']['class_1'])
        concept_five_class_odd.append(TCAV[ite]['Five']['class_1'])
        concept_six_class_odd.append(TCAV[ite]['Six']['class_1'])
        concept_seven_class_odd.append(TCAV[ite]['Seven']['class_1'])
        concept_eight_class_odd.append(TCAV[ite]['Eight']['class_1'])
        concept_nine_class_odd.append(TCAV[ite]['Nine']['class_1'])
    
    return concept_zero_class_even, concept_one_class_even, concept_two_class_even, concept_three_class_even, \
concept_four_class_even, concept_five_class_even, concept_six_class_even, concept_seven_class_even,\
concept_eight_class_even, concept_nine_class_even, \
concept_zero_class_odd, concept_one_class_odd, concept_two_class_odd, concept_three_class_odd, \
concept_four_class_odd, concept_five_class_odd, concept_six_class_odd, concept_seven_class_odd, \
concept_eight_class_odd, concept_nine_class_odd
```

```python
cav_flattening_type = "max_pooled"
max_pool_concept_zero_class_even, max_pool_concept_one_class_even, max_pool_concept_two_class_even, \
max_pool_concept_three_class_even, max_pool_concept_four_class_even, max_pool_concept_five_class_even,\
max_pool_concept_six_class_even, max_pool_concept_seven_class_even, max_pool_concept_eight_class_even, \
max_pool_concept_nine_class_even, \
max_pool_concept_zero_class_odd, max_pool_concept_one_class_odd, max_pool_concept_two_class_odd, \
max_pool_concept_three_class_odd, \
max_pool_concept_four_class_odd, max_pool_concept_five_class_odd, max_pool_concept_six_class_odd, \
max_pool_concept_seven_class_odd, \
max_pool_concept_eight_class_odd, max_pool_concept_nine_class_odd = get_TCAV(cav_flattening_type)
```

```python
def plot_bar_chart(
    flattening_c, 
    avg_c, 
    max_c, 
    title, 
    y_label, 
    x_label, 
    chart1_label="Flattened", 
    chart2_label="AvgPooled", 
    chart3_label="MaxPooled", 
    N=0
):
    import matplotlib.pyplot as plt
    import pandas as pd

    percent_weight_remaining = [
        "100 (1)",   "90 (2)",   "81 (3)",   "72.9 (4)",  "65.6 (5)",  "59.1 (6)",  
        "53.2 (7)",  "47.8 (8)",  "43.1 (9)",  "38.8 (10)",  "34.9 (11)",  "31.4 (12)",
        "28.3 (13)",  "25.4 (14)",  "22.9 (15)",  "20.6 (16)",  "18.6 (17)",  "16.7 (18)",  
        "15 (19)",   "13.5 (20)",  "12.2 (21)",  "11 (22)",    "9.9 (23)",   "8.9 (24)",
        "8 (25)",    "7.2 (26)",   "6.5 (27)",   "5.9 (28)",   "5.3 (29)",   "4.7 (30)",   
        "4.3 (31)",   "3.9 (32)",   "3.5 (33)",   "3.1 (34)",   "2.8 (35)",]

    plt.rcParams["figure.figsize"] = [20, 10]
    plt.rcParams["figure.autolayout"] = True

    d = {
        "X": percent_weight_remaining,
        chart1_label: flattening_c,
        chart2_label: avg_c,
        chart3_label: max_c
    }

    df = pd.DataFrame(d)
    df.set_index('X').plot(kind="bar", align='center', width=0.8)
    plt.tick_params(rotation=45)
    plt.title(f"{title} ({dataset_name}, {model_arch})")
    plt.xlabel('Unpruned weight percentage (iteration)')
    plt.ylabel('TCAV score')
    plt.show()
```

```python
def get_stats(cav_flattening_type, stat_file):
    prune_stat_path = os.path.join(
            logs,
            "predictions",
            "prune-statistics",
            model_arch,
            dataset_name,
            f"Prune_type_{prune_type}",
            f"cav_flattening_type_{cav_flattening_type}"
    )

    print(os.path.join(prune_stat_path, stat_file))
    prune_full_file = open(
            os.path.join(prune_stat_path, stat_file),
            "rb")
    prune_stats = pickle.load(prune_full_file)
    print(len(prune_stats))
    print(prune_stats)
    BB_acc = []
    BB_precision = []
    BB_recall = []
    BB_Roc_AUC = []
    BB_F1_score = []

    G_acc = []
    G_precision = []
    G_recall = []
    G_Roc_AUC = []
    G_F1_score = []

    completeness_score = []
    for idx, stat in enumerate(prune_stats):
        BB_acc.append(stat["BB"]["Accuracy"])
        BB_precision.append(stat["BB"]["Precision"])
        BB_recall.append(stat["BB"]["Recall"])
        BB_Roc_AUC.append(stat["BB"]["RocAUC"])
        BB_F1_score.append(stat["BB"]["F1_score"])

        G_acc.append(stat["G"]["Accuracy"])
        G_precision.append(stat["G"]["Precision"])
        G_recall.append(stat["G"]["Recall"])
        G_Roc_AUC.append(stat["G"]["RocAUC"])
        G_F1_score.append(stat["G"]["F1_score"])

        completeness_score.append(stat["Completeness_score"])
    
    return BB_acc, BB_precision, BB_recall, BB_Roc_AUC, BB_F1_score, G_acc, G_precision, G_recall, G_Roc_AUC, G_F1_score, completeness_score

def plot_graph(BB, percent_weight, title, y_label, x_label, y_lim):
    a = np.arange(len(BB))    
    plt.rcParams["figure.figsize"] = (20,10)
    plt.plot(a, np.array(BB), c="blue", label="Winning tickets") 
    plt.title(f"{'MNIST'} ({'ResNet18'})", fontdict={'fontsize': 20}) 
    plt.xlabel(x_label, fontsize=15) 
    plt.ylabel(y_label, fontsize=15) 
    plt.xticks(a, percent_weight, fontsize=12, rotation ="vertical") 
    plt.yticks(fontsize=12)
    plt.ylim(y_lim[0], y_lim[1])
    plt.grid(color="gray") 
    plt.savefig(f'/ocean/projects/asc170022p/shg121/PhD/Project_Pruning/plots/mnst_quant.png', dpi=220)
    # plt.savefig(f"{plot_path}/{prune_type}_val_Accuracy_Vs_Weights.png", dpi=1200) 
    plt.show()
    plt.close()

def plot_bar_chart(BB, G, title, y_label, x_label, chart1_label, chart2_label, N=0):
    if N ==0:
        N = len(percent_weight_remaining)
    fig, ax = plt.subplots(figsize=(20,10))

    ind = np.arange(N)    # the x locations for the groups
    width = 0.35         # the width of the bars
    ax.bar(ind, BB, width, label=chart1_label)
    ax.bar(ind + width, G, width,
           label=chart2_label)

    ax.set_title(f"{title} ({dataset_name}, {model_arch})")
    ax.set_xticks(ind + width / 2)
    if N == 0:
        ax.set_xticklabels(percent_weight_remaining, rotation=45)
    else:
        ax.set_xticklabels(percent_weight_remaining[0:N], rotation=45)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.legend()
    ax.autoscale_view()

    plt.show()
```
